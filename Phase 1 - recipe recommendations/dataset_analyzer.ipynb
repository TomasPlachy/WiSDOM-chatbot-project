```python
import json as json


def main():
    recipes_count = 0
    recipes_with = {
        "displayName" : 0,
        "description" : 0,
        "canonicalName": 0,
        "prepTime": 0,
        "cookTime": 0,
        "totalTime": 0,
        "method": 0,
        "difficultyLevel": 0,
        "images": 0,
        "videos":0,
        "tools": 0,
        "cuisines": 0,
        "meals": 0,
        "courses": 0,
        "occasions": 0,
        "diets": 0,
        "trimmed": 0,
        "difficulty": 0,
        "servings": 0,
        "components": 0,
        "meals": 0,
        "ingredients": 0,
        "instructions": 0,
        "nutrition": 0,
        "types": 0
    }
    
    with open("recipes_data.json", "r") as read_file:
        recipes_results = json.load(read_file)
    
        for key, recipe in recipes_results.items():
            recipes_count += 1
            if recipe['displayName'] is not None and recipe['displayName'] != '':
                recipes_with["displayName"] += 1
            if recipe['description'] is not None and recipe['description'] != '':
                recipes_with["description"] += 1
            if recipe['canonicalName'] is not None and recipe['canonicalName'] != '':
                recipes_with["canonicalName"] += 1
            if recipe['prepTimeMinutes'] is not None and recipe['prepTimeMinutes'] != '':
                recipes_with["prepTime"] += 1
            if recipe['cookTimeMinutes'] is not None and recipe['cookTimeMinutes'] != '':
                recipes_with["cookTime"] += 1
            if recipe['totalTimeMinutes'] is not None and recipe['totalTimeMinutes'] != '':
                recipes_with["totalTime"] += 1
            if recipe['cookingMethod'] is not None and recipe['cookingMethod'] != '':
                recipes_with["method"] += 1
            if recipe['difficultyLevel'] is not None and recipe['difficultyLevel'] != '':
                recipes_with["difficultyLevel"] += 1
            if recipe['images'] is not None and len(recipe['images']) != 0:
                recipes_with["images"] += 1
            if recipe['videos'] is not None and len(recipe['videos']) != 0:
                recipes_with["videos"] += 1
            if recipe['tools'] is not None and len(recipe['tools']) != 0:
                recipes_with["tools"] += 1
            if recipe['cuisines'] is not None and recipe['cuisines'] != '':
                recipes_with["cuisines"] += 1
            if recipe['meals'] is not None and len(recipe['meals']) != 0:
                recipes_with["meals"] += 1
                #print(recipe['meals'])
                #print(recipe['occasions'])
                #print("\n")
            if recipe['courses'] is not None and len(recipe['courses']) != 0:
                recipes_with["courses"] += 1
            if recipe['occasions'] is not None and len(recipe['occasions']) != 0:
                recipes_with["occasions"] += 1
            if recipe['diets'] is not None and len(recipe['diets']) != 0:
                recipes_with["diets"] += 1
            if recipe['trimmed'] is True:
                recipes_with["trimmed"] += 1
            if recipe['difficulty'] is not None and recipe['difficulty'] != '':
                recipes_with["difficulty"] += 1
            if recipe['servings'] is not None and recipe['servings'] != '':
                recipes_with["servings"] += 1
            if recipe['components'] is not None and len(recipe['components']) != 0 and recipe['components'][0]['name'] is not None:
                recipes_with["components"] += 1
            if recipe['ingredients'] is not None and len(recipe['ingredients']) != 0:
                recipes_with["ingredients"] += 1
            if recipe['instructions'] is not None and len(recipe['instructions']) != 0:
                recipes_with["instructions"] += 1
            if recipe['nutrition'] is not None and len(recipe['nutrition']) != 0 and (recipe['nutrition']['servingSizeQuantity'] is not None) :
                recipes_with["nutrition"] += 1
            if recipe['types'] is not None and len(recipe['types']) != 0:
                recipes_with["types"] += 1
                
        for key, value in recipes_with.items():
            print("Recipes with " + str(key) + ": " + str(value) + "/" + str(recipes_count))
        
    
main()
        
```
